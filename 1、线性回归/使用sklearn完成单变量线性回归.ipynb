```python
import numpy as np                     
import pandas as pd                    
import matplotlib.pyplot as plt  
from sklearn.linear_model import LinearRegression   #将模型导入线性回归中
from sklearn.model_selection import train_test_split
```

1. 导入数据集

```python
data = pd.read_csv('ex1data1.txt',names=['population','profit']) 
data.plot.scatter('population','profit',label='profit of population') # 绘制散点图
plt.show()  
```

2.构造数据集

```python
feature_columns = ['population']
label_column = ['profit']

features = data[feature_columns]
label = data[label_column]       # 定义特征列 和 标签列
```

```python
X = features.values  #X向量为特征
Y = label.values
X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size=0.3) 
```

3.使用sklearn中的线性回归算法训练数据

```python
# 训练数据
regressor = LinearRegression()
regressor = regressor.fit(X_train, Y_train)
```

```python
# 预测数据
Y_pred = regressor.predict(X_test)
```

```python
# 散点图：红色点表示训练集的点
plt.scatter(X_train , Y_train, color = 'red')
# 线图：蓝色线表示由训练集训练出的线性回归模型
plt.plot(X_train , regressor.predict(X_train), color ='blue')
plt.show()
```

```python
# 散点图：红色点表示测试集的点
plt.scatter(X_test , Y_test, color = 'red')
# 线图：蓝色线表示对测试集进行预测的结果
plt.plot(X_test ,Y_pred, color ='blue')
plt.show()
```
